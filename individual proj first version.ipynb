## finding the nearest police district to each location of violation 
Code obtained from : https://gis.stackexchange.com/questions/222315/geopandas-find-nearest-point-in-other-dataframe

```python
from scipy.spatial import cKDTree  
def ckdnearest(gdA, gdB, bcol):   
    nA = np.array(list(zip(gdA.LATITUDE, gdA.LONGITUDE)) )
    nB = np.array(list(zip(gdB.LATITUDE, gdB.LONGITUDE)) )
    btree = cKDTree(nB)
    dist, idx = btree.query(nA,k=1)
    df = pd.DataFrame.from_dict({'distance': dist.astype(int),
                             'bcol' : gdB.loc[idx, bcol].values })
    return df
```

```python
a=ckdnearest(df, df2,'DISTRICT NAME')
```

```python
a
```

## allocating the nearest distrcit to each record in the speed violations dataset. 

```python
df['cops']=a['bcol']
```

```python
df3=pd.merge(df,df2,how='left',left_on=df['cops'],right_on=df2['DISTRICT NAME'])
```
